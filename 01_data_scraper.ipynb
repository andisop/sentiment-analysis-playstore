```python
# Import library
import pandas as pd
from google_play_scraper import Sort, reviews
import os
```

**Scraping Data Reviews**

```python
def scrape_reviews(app_id, lang='id', country='id'):
    print(f'Scraping data reviews for app ID: {app_id}...')

    # Scraping reviews data Google Play API
    results, tokens = reviews(
        app_id,
        lang=lang,
        country=country,
        sort=Sort.MOST_RELEVANT,
        count=20000
    )

    df = pd.DataFrame(results)

    # Defining relative path
    directory = 'data'
    if not os.path.exists(directory):
        os.makedirs(directory)
        print(f"Directory '{directory}' created.")

    file_name = f"{app_id.split('.')[1]}_app_reviews.csv"
    file_path = os.path.join(directory, file_name)

    # Save dataset
    df.to_csv(file_path, index=False)

    print(f"{len(df)} reviews saved to {file_path}")
    return df
```

```python
# Scraping reviews for aplication SIGNAL–SAMSAT DIGITAL NASIONAL
app_id_target = 'app.signal.id'
df_reviews = scrape_reviews(app_id_target)
```
